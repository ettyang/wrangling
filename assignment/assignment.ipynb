```python
### PROBLEM 2 ###
df = pd.read_csv('./wrangling/assignment/data/sharks.csv', low_memory=False)
df.head()
```

```python
type = df['Type']
type.value_counts()
```

```python
type = type.replace(['Sea Disaster','Boat','Boatomg', 'Boating'],'Watercraft') #put all boating in 'watercraft
type.value_counts()
type = type.replace(['Invalid','Questionable','Unconfirmed','Unverified', 'Under investigation'],np.nan)#put all uncategorized values
type.value_counts()

df['Type'] = type
del type

df['Type'].value_counts()
```

```python
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace('y','Y') #standardize captalization
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace(['UNKNOWN','F','M','2017'],np.nan) #taking out unknown or irrelevant stuff
pd.crosstab(df['Type'],df['Fatal (Y/N)'],normalize='index') #break variables apart

```

```python
### PROBLEM 3 ###
url = 'http://www.vcsc.virginia.gov/pretrialdataproject/October%202017%20Cohort_Virginia%20Pretrial%20Data%20Project_Deidentified%20FINAL%20Update_10272021.csv'
df = pd.read_csv(url,low_memory=False)
```

```python
pretrial = df['WhetherDefendantWasReleasedPretrial']
print(pretrial.value_counts(),'\n')
print(pretrial.unique(),'\n')

pretrial = pretrial.replace(9,np.nan)
sum(pretrial.isnull())
df['WhetherDefendantWasReleasedPretrial'] = pretrial
del pretrial
```

```python
### PROBLEM 4 ###
imposed = df['ImposedSentenceAllChargeInContactEvent']
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']

#print(imposed.unique(), '\n') #----has some empty values
#df['ImposedSentenceAllChargeInContactEvent'].dtypes #---shows column type = string
imposed = pd.to_numeric(imposed, errors='coerce') #coerce variables to numeric
imposed_NA = imposed.isnull() #missing dummy
print(sum(imposed_NA),'\n')#shows 9053 missing values
```

```python
type.unique() # type 4 cases were dismissed
print(pd.crosstab(type, imposed_NA), '\n') # type 4 cases were dismissed

#replace type 4 cases w/ imposed = 0, replace type 9 cases w/ imposed = nan
imposed = imposed.mask(type == 4, 0)
imposed = imposed.mask(type == 9, np.nan)

#make missing dummy and find how many are still missing
imposed_NA = imposed.isnull()
print(np.sum(imposed_NA),'\n') #only 274 missing now!!!

#replace clean data into column and delete imposed variable
df['ImposedSentenceAllChargeInContactEvent'] = imposed
del imposed
```
